```python
# Connect to the 'orders' database
mydb = pymysql.connect(
    host="localhost",
    user="root",
    passwd="Ankit@123",
    database="orders"
)

# Create a cursor object
cursor = mydb.cursor()

# SQL command to create a table
create_table_query = """
CREATE TABLE IF NOT EXISTS df_orders (
    order_id INT PRIMARY KEY,
    order_date DATE,
    ship_mode VARCHAR(20),
    segment VARCHAR(20),
    country VARCHAR(20),
    city VARCHAR(20),
    state VARCHAR(20),
    postal_code VARCHAR(20),
    region VARCHAR(20),
    category VARCHAR(20),
    sub_category VARCHAR(20),
    product_id VARCHAR(50),
    quantity INT,
    discount DECIMAL(7,2),
    sale_price DECIMAL(7,2),
    profit DECIMAL(7,2)
)
"""

# Execute the SQL command to create the table
cursor.execute(create_table_query)

# Close the cursor and connection
cursor.close()
mydb.close()

print("Table 'df_orders' created successfully")
```

```python
import pandas as pd
import pymysql

# Assuming 'df' is your DataFrame containing the data to be inserted

# Establish connection to MySQL server
mydb = pymysql.connect(
    host="localhost",
    user="root",
    passwd="Ankit@123",
    database="orders"
)

# Create a cursor object
cursor = mydb.cursor()

# Example: Insert data from DataFrame into MySQL table
try:
    # Iterate over each row in the DataFrame
    for index, row in df.iterrows():
        # Prepare data tuple
        data = (
            row['order_id'],
            row['order_date'],
            row['ship_mode'],
            row['segment'],
            row['country'],
            row['city'],
            row['state'],
            row['postal_code'],
            row['region'],
            row['category'],
            row['sub_category'],
            row['product_id'],
            row['quantity'],
            row['discount'],
            row['sale_price'],
            row['profit']
        )

        # SQL query to insert data into df_orders table
        insert_query = """
        INSERT INTO df_orders (order_id, order_date, ship_mode, segment, country, city, state, postal_code, region, category, sub_category, product_id, quantity, discount, sale_price, profit)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        """

        # Execute the SQL query
        cursor.execute(insert_query, data)

    # Commit changes to the database
    mydb.commit()
    print("Data inserted successfully into 'df_orders' table")

except pymysql.Error as e:
    print(f"Error: {e}")
    # Rollback in case of error
    mydb.rollback()

finally:
    # Close cursor and connection
    cursor.close()
    mydb.close()
```
